## Part 10 - Création de features supplémentaires test

```python
import pandas as pd
import spacy
```

```python
# Chargement de la dataset
dataset = pd.read_csv("dataset_CV_labelise_features1_200.csv", delimiter = ";", encoding = "utf-8")
len(dataset)
```

```python
dataset_sample = dataset.head(20)
```

```python
dataset_sample.head()
```

```python
nlp = spacy.load("fr_core_news_lg")
```

```python
dataset_sample["Tokenization"] = dataset_sample["Sentences_CV_clean"].apply(lambda x : nlp(x))
```

```python
def count_verb(text, nlp):
    increment  = 0
    tokens = nlp(text)
    for token in tokens:
        if token.pos_ == 'VERB':
            increment = increment + 1
    return increment
```

```python
def count_noun(text, nlp):
    increment  = 0
    tokens = nlp(text)
    for token in tokens:
        if token.pos_ == 'NOUN':
            increment = increment + 1
    return increment
```

```python
def count_propn(text, nlp):
    increment  = 0
    tokens = nlp(text)
    for token in tokens:
        if token.pos_ == 'PROPN':
            increment = increment + 1
    return increment
```

```python
def count_num(text, nlp):
    increment  = 0
    tokens = nlp(text)
    for token in tokens:
        if token.pos_ == 'NUM':
            increment = increment + 1
    return increment
```

```python
dataset_sample["Verb_count"] = dataset_sample["Tokenization"].apply(lambda x : count_verb(x, nlp))
dataset_sample["Propn_count"] = dataset_sample["Tokenization"].apply(lambda x : count_propn(x, nlp))
dataset_sample["Noun_count"] = dataset_sample["Tokenization"].apply(lambda x : count_noun(x, nlp))
dataset_sample["Num_count"] = dataset_sample["Tokenization"].apply(lambda x : count_num(x, nlp))
```

```python
dataset_sample["Pourcentage_verb_sentence"] = dataset_sample["Verb_count"] / dataset_sample["Nb_tokens"]
dataset_sample["Pourcentage_propn_sentence"] = dataset_sample["Propn_count"] / dataset_sample["Nb_tokens"]
dataset_sample["Pourcentage_noun_sentence"] = dataset_sample["Noun_count"] / dataset_sample["Nb_tokens"]
dataset_sample["Pourcentage_num_sentence"] = dataset_sample["Num_count"] / dataset_sample["Nb_tokens"]
```
